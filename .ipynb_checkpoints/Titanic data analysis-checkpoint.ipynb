```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
```

```python
df = pd.read_csv('titanic_data.csv')
# df.head(3)
df.rename(columns={'Pclass':'Ticket_class','Parch':'ParCh','Embarked':'Departure', 'Ticket':'Ticket_No'},inplace=True)
```

```python
df.info()
```

#### How old was who were survived ?

```python
# df[df['Survived'] == 1]['Age'].hist();
ax = df[df['Survived'] == 1].hist('Age')
ax = ax[0]
for x in ax:
    # Set title
    x.set_title('How old was who were survived ?', size=15)
    # Set x-axis label
    x.set_xlabel("Age",labelpad=5,  size=15)

    # Set y-axis label
    x.set_ylabel("Passengers", labelpad=5, size=15)
```

#### How many was who had siblings or spouses aboard the Titanic ?

```python
ax = df.hist(column='SibSp')

ax = ax[0]
for x in ax:
    # Set title
    x.set_title('How many was who had siblings or spouses \n aboard the Titanic ?', size=15)
    # Set x-axis label
    x.set_xlabel("No. of siblings / spouses",labelpad=5,  size=15)

    # Set y-axis label
    x.set_ylabel("Passengers", labelpad=5, size=15)
```

#### How many was who had  parents / children aboard the Titanic ?

```python
ax = df.hist(column='ParCh')

ax = ax[0]
for x in ax:
    # Set title
    x.set_title('How many was who had parents / children \n aboard the Titanic ?',\
                size=15)
    # Set x-axis label
    x.set_xlabel("No. of parents / children",labelpad=5,  size=15)

    # Set y-axis label
    x.set_ylabel("Passengers", labelpad=5, size=15)
```

#### What is the total number of each gender aboard the Titanic ?

```python
sex_count = dict()
male = df[df['Sex'] == 'male']['Sex'].count()
female = df[df['Sex'] == 'female']['Sex'].count()

sex_count['Passengers'] = np.array((male, female))
sex_df = pd.DataFrame(data = sex_count, index=['Male', 'Female'])
sex_df
title = 'What is the total number of each gender \n aboard the Titanic ?'
ax = sex_df.plot(kind='bar',rot=0)
ax.set_title(title, size=15);
ax.set_xlabel('Gender', size=15, labelpad=5)
ax.set_ylabel('Passengers', size=15, labelpad=5);
```

#### What was the departure port of each who were survived aboard the Titanic?

```python
departure = dict()
departure['Southampton'] = [df[(df['Survived'] == 1) & (df['Departure'] == 'S')]\
['PassengerId'].count(), df[df['Departure'] == 'S']\
['PassengerId'].count()]
departure['Cherbourg'] = [df[(df['Survived'] == 1) & (df['Departure'] == 'C')]\
['PassengerId'].count(), df[df['Departure'] == 'C']\
['PassengerId'].count()]
departure['Queenstown'] = [df[(df['Survived'] == 1) & (df['Departure'] == 'Q')]\
['PassengerId'].count(), df[df['Departure'] == 'Q']\
['PassengerId'].count()]
data = departure.values()
index = departure.keys()
ax = pd.DataFrame(data=data, index=index, columns=['Survived', 'Total']).plot(kind='bar',rot=0)
ax.set_title('What was the departure port \
of each who were \n survived aboard the Titanic?'\
             , size=15);
ax.set_xlabel('Cities', size=15, labelpad=10);
ax.set_ylabel('Passengers', size=15, labelpad=10);
```

```python
departure = dict()
departure['Southampton'] = [df[(df['Survived'] == 0) & (df['Departure'] == 'S')]\
['PassengerId'].count(), df[df['Departure'] == 'S']\
['PassengerId'].count()]
departure['Cherbourg'] = [df[(df['Survived'] == 0) & (df['Departure'] == 'C')]\
['PassengerId'].count(), df[df['Departure'] == 'C']\
['PassengerId'].count()]
departure['Queenstown'] = [df[(df['Survived'] == 0) & (df['Departure'] == 'Q')]\
['PassengerId'].count(), df[df['Departure'] == 'Q']\
['PassengerId'].count()]
data = departure.values()
index = departure.keys()
ax = pd.DataFrame(data=data, index=index, columns=['Not Survived', 'Total']).plot(kind='bar',rot=0)
ax.set_title('What was the departure port \
of each who were \n survived aboard the Titanic?'\
             , size=15);
ax.set_xlabel('Cities', size=15, labelpad=10);
ax.set_ylabel('Passengers', size=15, labelpad=10);
```

#### What was the ticket class of each who were survived aboard the Titanic ?


```python
ticket_class = dict()
ticket_class['First class'] = [df[(df['Survived'] == 1) & (df['Ticket_class'] == 1)]\
['PassengerId'].count(), df[df['Ticket_class'] == 1]\
['PassengerId'].count()]
ticket_class['Second class'] = [df[(df['Survived'] == 1) & (df['Ticket_class'] == 2)]\
['PassengerId'].count(), df[df['Ticket_class'] == 2]\
['PassengerId'].count()]
ticket_class['Third class'] = [df[(df['Survived'] == 1) & (df['Ticket_class'] == 3)]\
['PassengerId'].count(), df[df['Ticket_class'] == 3]\
['PassengerId'].count()]
data = ticket_class.values()
index = ticket_class.keys()
ax = pd.DataFrame(data=data, index=index, columns=['Survived', 'Total']).plot(kind='bar',rot=0)
ax.set_title('What was the ticket class \
of each who were \n survived aboard the Titanic ?'\
             , size=15);
ax.set_xlabel('Ticket class', size=15, labelpad=10);
ax.set_ylabel('Passengers', size=15, labelpad=10);
```
